```python
def check(x,y,k):
    unique=[int(x) for x in set(str(x))]
    if len(unique)>y:
        k=k+1
        return check(x+1,y,k)
    else:
        return k
    return k    
n=int(input())
for i in range(n):
    x,y=map(int,input().split())
    z=check(x,y,0)
    print(z)
```

```python
import sys
sys.setrecursionlimit(10**7)
def count_distinct_digits(n):
    return len(set(str(n)))

def find_minimum_x(n, k):
    x = 0
    while True:
        if count_distinct_digits(n + x) <= k:
            return x
        x += 1
find_minimum_x(90032975,1)

```
