```python
import numpy as np

np.random.seed(42)
df = pd.DataFrame({"Google": np.random.randn(1000)+0.2, 
                   "Apple": np.random.randn(1000)+0.17}, 
                   index=pd.date_range('1/1/2000', periods=1000))
df = df.cumsum()
df = df + 50
df.plot_bokeh(kind="line")
```

```python
df.plot_bokeh(kind='line',xlabel='date',ylabel='stock price$',ylim=(0,400),title='apple vs google',colormap=['red','green'])
```

```python
df.plot_bokeh(kind='line',title='apple vs google',xlabel='date',ylabel='stockprice$',figsize=(800,450),ylim=(100,200),xlim=('2001-01-01','2001-02-01'),plot_data_points=True,plot_data_points_size=10,marker='asterisk',toolbar_location='left')
```

```python
x=np.arange(-3,3,0.1)
y2=x**2
y3=x**3
df=pd.DataFrame({'x':x,'parabola':y2,'cubic':y3})
df.plot_bokeh(kind='point',x='x',xticks=range(-3,4),size=5,colormap=['red','green'],title='parabola vs cube',marker='square')
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris["data"])
df.columns = iris["feature_names"]
df["species"] = iris["target"]
df["species"] = df["species"].map(dict(zip(range(3), iris["target_names"])))
df.plot_bokeh(kind='scatter',x='petal length (cm)',y='sepal width (cm)',category='species',show_figure=True)
```

```python
data={'fruits':['apple','pears','mango','grapes','guava'],
     '2015':[2,4,6,1,8],
     '2016':[4,1,6,4,4],
     '2017':[8,4,3,2,4]}
df=pd.DataFrame(data).set_index('fruits')
df.plot_bokeh(kind='bar',ylabel='price per unit$',title='fruits price per year',alpha=0.8)

```

```python
df_hist=pd.DataFrame({'a':np.random.randn(1000)+1,
                    'b':np.random.randn(1000),
                     'c':np.random.randn(1000)-1},columns=['a','b','c'])
#default top on top histogram
df_hist.plot_bokeh(kind='hist',bins=np.linspace(-5,5,41),vertical_xlabel=True,hovertool=False,title='Normal distributions(top on top)',line_color='black')
#side by side histogram multiple bars share bins side by side
df_hist.plot_bokeh(kind='hist',bins=np.linspace(-5,5,41),histogram_type='sidebyside',vertical_xlabel=True,hovertool=False,title='normal distributions(side by side)',line_color='black')
#stacked histogram
df_hist.plot_bokeh(kind='hist',bins=np.linspace(-5,5,41),histogram_type='stacked',vertical_xlabel=True,hovertool=False,title='normal distribution(stacked)',line_color='black')
```
