```python
import numpy as np
import pandas as pd
from matplotlib import pylab
import matplotlib.pyplot as plt
import math
```

```python
x=np.linspace(0,10,25)
y=x*x+2
print(x)
print(y)
```

```python
pylab.plot(x,y,'r--o')
```

```python
a=np.sin(x)
b=np.cos(x)
```

```python
pylab.subplot(2,2,1)
pylab.plot(x,y)
pylab.subplot(2,2,2)
pylab.plot(y,x)
pylab.subplot(2,2,3)
pylab.plot(x,a)
pylab.subplot(2,2,4)
pylab.plot(x,b)
```

```python
fig=plt.figure()
axis=fig.add_axes([0.1,0.5,0.8,0.8])
axis.plot(x,y,'r')
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1)
for ax in axes:
    ax.plot(x,y,'r')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.6,0.2,0.2])
axes1.plot(x,y,'r')
axes2.plot(y,x,'g')
```

```python
fig=plt.figure(figsize=(16,9),dpi=100)
plt.plot(x,y,'r')

```

```python
fig,axes=plt.subplots(figsize=(16,9),dpi=200)
axes.plot(x,y,'r')
```

```python
fig,axes=plt.subplots()
axes.set_title('title')
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.plot(x,x**2,'r')
axes.plot(x,x**3,'g')
axes.legend(['y=x**2','y=x**3'],loc=2)
```

```python
fig,axes=plt.subplots(dpi=300)
axes.plot(x,x+1,'b',alpha=1)
axes.plot(x,x+2,'g',alpha=0.6)
axes.plot(x,x**2,'r',alpha=0.8)
```

```python
fig,axes=plt.subplots(dpi=300)
axes.plot(x,x+2,color='green',linewidth=0.25,linestyle='--')
axes.plot(x,x+3,color='green',linewidth=1,linestyle='-.')
axes.plot(x,x+4,color='green',lw=1.5,ls='dashed')
axes.plot(x,x+5,color='green',lw=2,ls='dashdot')
line,=axes.plot(x,x+7,color='black',lw=2)
line.set_dashes([5,10,10,5])
```

```python
fig,axes=plt.subplots(dpi=300)
axes.plot(x,x+1,marker='o',markersize=3)
axes.plot(x,x+2,marker='+')
axes.plot(x,x+3,marker='s',markersize=8,markerfacecolor='r')
```

```python
fig,axes=plt.subplots(1,2,figsize=(10,5),dpi=800)
#axes[0].plot(x,x**2,lw=2)
#axes[0].plot(x,x**3,lw=2)
axes[0].plot(x,x**2,x,x**3,lw=2)
axes[0].grid(True)
axes[1].plot(x,x**2,x,x**3,lw=2)
axes[1].set_ylim([0,60])
axes[1].set_xlim([2,5])
```

```python
fig,axes=plt.subplots(1,4,figsize=(16,5))
axes[0].set_title('scatter plot')
axes[0].scatter(x,x+0.25*np.random.rand(len(x)))
n=np.array([0,1,2,3,4,5])
axes[1].set_title("step plot")
axes[1].step(n,n**2)
axes[2].set_title('bar plot')
axes[2].bar(n,n**2,align='center',alpha=1,width=0.8)
axes[3].set_title('fill between')
axes[3].fill_between(x,x**2,x**3,color='green',alpha=0.8)
```

```python
fig=plt.figure(figsize=(6,6))
ax=fig.add_axes([0.1,0.1,1,1],polar=True)
t=np.linspace(0,2*math.pi,100)
ax.plot(t,t,color='red',lw=3)
ax.plot(t,t*2,color='green',lw=3)
ax.plot(t,t*3,color='yellow',lw=3)
```

```python
n=np.random.randn(1000)
fig,axes=plt.subplots(1,3,figsize=(12,4))
axes[0].set_title('default histogram')
axes[0].hist(n)
axes[1].set_title('cumulative histogram')
axes[2].set_title('cumulative histogram')
axes[1].hist(n,cumulative=True,bins=10)
axes[2].hist(n,bins=10)
```

# practical examples

```python
import matplotlib.pyplot as plt
languages=['Python','Java','C','C++','PHP','Javascript']
popularity=[22.2,17.6,8.8,8,7.7,6.7]
colors=['red','green','blue','yellow','pink','orange']
```

```python
explode=(0.1,0,0,0,0,0)
plt.pie(popularity,explode=explode,labels=languages,colors=colors,autopct='%1.1f%%',shadow=True,startangle=180)
```
